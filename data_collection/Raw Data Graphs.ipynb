```python
import matplotlib.pyplot as plt
import numpy as np

x_acc_threshold = []
y_acc_threshold = []
z_acc_threshold = []

x_acc = []
y_acc = []
z_acc = []

i = 0
colors= ['b-', 'g-', 'r-', 'c-', 'm-']
file = open("outwardpush.txt", "r")
threshold = 0.5
f, axarr = plt.subplots(3, sharex=True)
f_threshold, axarr_threshold = plt.subplots(3, sharex=True)

for line in file:
    if "PUSHED" not in line:
        x = float(line[:line.find(',')])
        y = float(line[line.find(',')+1:line.rfind(',')])
        z = float(line[line.rfind(',')+1:])
        if x > threshold or x < -1*threshold:
            x_acc_threshold.append(x)
        else:
            x_acc_threshold.append(0.0)
        if y > threshold or y < -1*threshold:
            y_acc_threshold.append(y)
        else:
            y_acc_threshold.append(0.0)
        if z > threshold or z < -1*threshold:
            z_acc_threshold.append(z)
        else:
            z_acc_threshold.append(0.0)
        
        x_acc.append(x)
        y_acc.append(y)
        z_acc.append(z)
        
        
    elif "PUSHED" in line and "---" in line:
        #print(line)
        #print(i)
        #print(x_acc)
        axarr[0].plot(range(len(x_acc)), x_acc, colors[i])
        axarr[1].plot(range(len(y_acc)), y_acc, colors[i])
        axarr[2].plot(range(len(z_acc)), z_acc, colors[i])
        #print(x_acc_threshold)
        axarr_threshold[0].plot(range(len(x_acc_threshold)), x_acc_threshold, colors[i])
        axarr_threshold[1].plot(range(len(y_acc_threshold)), y_acc_threshold, colors[i])
        axarr_threshold[2].plot(range(len(z_acc_threshold)), z_acc_threshold, colors[i])
        i+=1
        x_acc = []
        y_acc = []
        z_acc = []
        x_acc_threshold = []
        y_acc_threshold = []
        z_acc_threshold = []
    

#axarr[0].plot(range(len(x_acc)), x_acc)  
#axarr[1].plot(range(len(y_acc)), y_acc)  
#axarr[2].plot(range(len(z_acc)), z_acc)  
axarr[0].set_title('Outward push acceleration (ax, ay, az)')
axarr_threshold[0].set_title('Outward push threshold acceleration (ax, ay, az)')
  
```
